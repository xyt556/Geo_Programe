# STAC 简介

**STAC（SpatioTemporal Asset Catalog）** 是一种用于组织、描述和发现时空数据（尤其是遥感影像数据）的开放标准。它最初由遥感与云计算社区发起，旨在解决海量地球观测数据在云环境中的共享、检索和互操作问题。

可以把 STAC 理解为：

> **地理空间数据领域的“数据目录标准”**，类似于图书馆中的图书编目系统。

目前 STAC 已成为云原生地理空间（Cloud-Native Geospatial）生态的重要基础标准，被广泛应用于：

* 遥感影像管理
* 地学大数据平台
* 数据共享门户
* AI训练数据集管理
* 云计算GIS平台

例如：

* Microsoft Planetary Computer
* [Earth Search](https://earth-search.aws.element84.com/?utm_source=chatgpt.com)
* Radiant Earth Foundation
* [Google Earth Engine Catalog Explorer](https://developers.google.com/earth-engine/datasets?utm_source=chatgpt.com)

都已采用或支持 STAC 标准。

---

# 为什么需要 STAC？

传统遥感数据通常存在以下问题：

Landsat/
├── LC08_20200501.tif
├── LC08_20200517.tif
├── metadata.xml
└── ...

不同机构：

* 元数据格式不同
* 文件命名不同
* 检索接口不同

导致：

* 数据难发现
* 难自动化处理
* 难跨平台共享

STAC通过统一的数据描述规范解决这一问题。

# STAC核心思想

STAC采用 JSON 格式描述数据。

一个遥感影像不仅包含文件本身：

```json
{
  "id": "LC08_20200501",
  "datetime": "2020-05-01T10:30:00Z",
  "bbox": [...],
  "geometry": {...},
  "assets": {...}
}
```

还包含：

* 时间
* 空间范围
* 波段信息
* 云量
* 数据来源
* 文件链接

使数据能够被机器自动发现和理解。

# STAC的数据层次结构

STAC采用三级结构：

```text
Catalog
│
├── Collection
│   ├── Item
│   ├── Item
│   └── Item
│
└── Collection
    ├── Item
    └── Item
```

## 1 Catalog（目录）

最高层组织结构。

例如：

```text
Planetary Computer
```

包含多个数据集。

---

## 2 Collection（数据集）

类似数据库中的表。

例如：

* Landsat-8
* Sentinel-2
* MODIS

每个 Collection 描述一类数据。

示例：

```json
{
  "id": "sentinel-2-l2a",
  "title": "Sentinel-2 Level-2A"
}
```

---

## 3 Item（数据条目）

对应单景影像。

例如：

```text
S2A_20240501_T49QCF
```

包含：

* 时间
* 空间位置
* 元数据

示例：

```json
{
  "id": "S2A_20240501",
  "datetime": "2024-05-01T03:12:21Z"
}
```

---

## 4 Asset（资源）

Item中的实际文件。

例如：

```json
{
  "B04": {
    "href": "s3://.../B04.tif"
  },
  "B08": {
    "href": "s3://.../B08.tif"
  }
}
```

对应：

* 红光波段
* 近红外波段

以及：

* Thumbnail
* Metadata
* QA文件

等。

# STAC API

随着数据量增加，仅靠静态JSON文件已不够。

因此发展出：

## STAC API

类似：

```http
GET /collections

GET /collections/sentinel-2/items

GET /search
```

查询示例：

```json
{
  "collections": ["sentinel-2"],
  "bbox": [116,39,117,40],
  "datetime": "2024-01-01/2024-12-31"
}
```

返回符合条件的影像列表。

# STAC与GeoJSON的关系

STAC实际上构建于 GeoJSON 之上。

普通GeoJSON：

```json
{
  "type": "Feature"
}
```

STAC Item：

```json
{
  "type": "Feature",
  "geometry": {...},
  "properties": {...},
  "assets": {...}
}
```

即：

> STAC = GeoJSON + 遥感元数据 + 数据链接

# pySTAC 示例

安装：

```bash
pip install pystac
```

读取 Catalog：

````python
import pystac

catalog = pystac.Catalog.from_file(
    "catalog.json"
)

for collection in catalog.get_children():
    print(collection.id)
````

---

# pystac-client 示例

连接 STAC API：

````python
from pystac_client import Client

catalog = Client.open(
    "https://planetarycomputer.microsoft.com/api/stac/v1"
)
````

搜索 Sentinel-2：

````python
search = catalog.search(
    collections=["sentinel-2-l2a"],
    bbox=[116,39,117,40],
    datetime="2024-01-01/2024-12-31"
)

items = list(search.items())
````

# 与地学AI的关系

STAC已经成为 GeoAI 数据基础设施的重要组成部分。

典型流程：

```text
STAC API
    ↓
pystac-client
    ↓
stackstac
    ↓
xarray
    ↓
PyTorch
    ↓
GeoAI
```

例如：

* 土地覆盖分类
* 作物识别
* 洪水监测
* 城市扩张检测
* 遥感大模型训练

都可以直接通过 STAC 获取训练样本。

# STAC + COG + GeoParquet

当前云原生地理空间领域的主流技术栈：

```text
             GeoAI
               ↑
          STAC API
               ↑
      ┌──────────────┐
      │     STAC     │
      └──────────────┘
          ↑      ↑
        COG   GeoParquet
```

其中：

* **STAC**：数据目录与检索标准
* **COG（Cloud Optimized GeoTIFF）**：云原生栅格格式
* **GeoParquet**：云原生矢量格式

三者共同构成现代地学大数据平台的核心基础设施。
