# Ipyleaflet


## 简介
**ipyleaflet**是一个Python库，将**Leaflet.js**的强大功能带到**Jupyter Notebooks**中。它允许使用小部件进行交互式地图绘制、地理空间可视化和动态地图定制。

`ipyleaflet` 是一个基于 **Leaflet.js** 的 Python 交互式地图库，专为 **Jupyter Notebook** 和 **JupyterLab** 设计。与 Folium 不同，它是一个真正的 **Jupyter Widget**，支持 **双向交互**，可以动态更新地图并获取用户操作反馈。

### 二、基本用法



````python

from ipyleaflet import Map, Marker

# 创建地图
m = Map(
    center=(39.9042, 116.4074),  # 中心点（纬度，经度）
    zoom=10                       # 缩放级别
)

# 在 Jupyter 中显示
m
````



直接输出变量 `m` 即可在 Notebook 中显示交互式地图。

### 三、常用参数

| 参数 | 说明 | 默认值 |
|------|------|--------|
| `center` | 地图中心，格式 `(纬度, 经度)` | `(0, 0)` |
| `zoom` | 缩放级别（1-18） | `10` |
| `basemap` | 底图样式 | `basemaps.OpenStreetMap.Mapnik` |
| `scroll_wheel_zoom` | 鼠标滚轮缩放 | `True` |
| `dragging` | 是否可拖动 | `True` |
| `layout` | 地图尺寸布局 | `Layout(height='600px')` |

示例：



````python

from ipyleaflet import Map, basemaps
from ipywidgets import Layout

m = Map(
    center=(31.2304, 121.4737),
    zoom=12,
    basemap=basemaps.Esri.WorldImagery,  # 卫星底图
    scroll_wheel_zoom=True,
    layout=Layout(width='100%', height='500px')
)
m
````

### 五、添加标记与图层

#### 1️⃣ 添加标记（Marker）



````python

from ipyleaflet import Marker

marker = Marker(location=(31.2304, 121.4737), draggable=True)
m.add_layer(marker)

# 监听标记拖动事件
def on_move(**kwargs):
    print(f"新位置: {kwargs['location']}")

marker.on_move(on_move)
m


````


#### 2️⃣ 添加圆形（Circle）



````python

from ipyleaflet import Circle

circle = Circle(
    location=(31.2304, 121.4737),
    radius=5000,  # 半径（米）
    color="red",
    fill_color="red",
    fill_opacity=0.3
)
m.add_layer(circle)
m
````


---

#### 3️⃣ 添加多边形（Polygon）



````python

from ipyleaflet import Polygon

polygon = Polygon(
    locations=[
        (30, 120),
        (31, 121),
        (30, 121)
    ],
    color="green",
    fill_color="green"
)
m.add_layer(polygon)
m
````



---

#### 4️⃣ 添加折线（Polyline）


````python


from ipyleaflet import Polyline

line = Polyline(
    locations=[(30, 120), (31, 121), (32, 122)],
    color="blue",
    weight=3
)
m.add_layer(line)
m
````

### 十、ipyleaflet vs Folium

| 特性 | ipyleaflet | Folium |
|------|-----------|--------|
| **交互性** | ✅ 双向交互（Widget） | ⚠️ 单向生成 HTML |
| **事件监听** | ✅ 支持 Python 回调 | ❌ 不支持 |
| **动态更新** | ✅ 实时修改地图 | ❌ 需重新生成 |
| **适用环境** | Jupyter 专用 | 任何 Python 环境 |
| **输出格式** | Jupyter Widget | HTML 文件 |
| **学习曲线** | 稍陡 | 简单 |

### 十一、完整示例


````python
from ipyleaflet import Map, Marker, Popup, basemaps
from ipywidgets import HTML

# 创建地图
m = Map(center=(39.9042, 116.4074), zoom=1, basemap=basemaps.Esri.WorldImagery)

# 添加标记
marker = Marker(location=(39.9042, 116.4074))
m.add_layer(marker)

# 添加弹出窗口
message = HTML(value="<b>北京</b><br>中国首都")
popup = Popup(child=message, close_button=False, auto_close=False)
marker.popup = popup

# 监听点击（更安全写法）
def on_click(**kwargs):
    if kwargs.get("type") == "click":
        print("点击坐标:", kwargs.get("coordinates"))

m.on_interaction(on_click)

m
````

## 5. 添加GeoJSON图层
GeoJSON文件允许可视化地理空间边界。

````python
from ipyleaflet import GeoJSON
import json

data = {
    "type": "FeatureCollection",
    "features": [
        {
            "type": "Feature",
            "geometry": {"type": "Point", "coordinates": [-122.4194, 37.7749]},
            "properties": {"name": "San Francisco"},
        }
    ],
}

gj = GeoJSON(data=data)
m.add_layer(gj)
m
````

## 6. 添加等值区域图层
等值区域地图可视化地理数据分布。

````python
import os
import json
import random
import requests

from ipyleaflet import Map, GeoJSON

if not os.path.exists("europe_110.geo.json"):
    url = "https://github.com/jupyter-widgets/ipyleaflet/raw/master/examples/europe_110.geo.json"
    r = requests.get(url)
    with open("europe_110.geo.json", "w") as f:
        f.write(r.content.decode("utf-8"))

with open("europe_110.geo.json", "r") as f:
    data = json.load(f)


def random_color(feature):
    return {
        "color": "black",
        "fillColor": random.choice(["red", "yellow", "green", "orange"]),
    }


m = Map(center=(50.6252978589571, 0.34580993652344), zoom=3)

geo_json = GeoJSON(
    data=data,
    style={"opacity": 1, "dashArray": "9", "fillOpacity": 0.1, "weight": 1},
    hover_style={"color": "white", "dashArray": "0", "fillOpacity": 0.5},
    style_callback=random_color,
)
m.add(geo_json)

m
````

## 7. 添加图层控制
动态切换图层：

````python
from ipyleaflet import LayersControl

m.add_control(LayersControl())
m
````

## 8. 添加交互式小部件
小部件实现与地图的实时交互。

````python
from ipywidgets import FloatSlider

zoom_slider = FloatSlider(description="Zoom Level", min=1, max=20, value=m.zoom)


def zoom_change(change):
    m.zoom = change["new"]


zoom_slider.observe(zoom_change, "value")
zoom_slider
````

## 11. 添加WMS图层

````python
from ipyleaflet import Map, WMSLayer, basemaps
````

````python
wms = WMSLayer(
    url="http://mesonet.agron.iastate.edu/cgi-bin/wms/nexrad/n0r.cgi",
    layers="nexrad-n0r-900913",
    format="image/png",
    transparent=True,
    attribution="Weather data © 2012 IEM Nexrad",
)

m = Map(basemap=basemaps.CartoDB.Positron, center=(38.491, -95.712), zoom=4)
m.add(wms)
m
````
