# 入门指南

## 简介

既然我们已经掌握了Python编程的基础知识，我们将开始探索使用Python进行地理空间数据分析和可视化。本章将向您介绍构成Python空间数据处理基础的关键地理空间库。

地理空间数据分析对于各种应用至关重要，包括环境监测、城市规划和地图制作。Python生态系统提供了一套强大的库来处理矢量和栅格数据、执行空间分析以及创建交互式地图。我们将使用的一些核心库包括：

- **[GeoPandas](https://geopandas.org)：**用于处理矢量数据（如shapefile、GeoJSON）并执行空间操作。
- **[Rasterio](https://rasterio.readthedocs.io)：**用于读取、分析和写入栅格数据（如卫星图像）。
- **[Xarray](https://xarray.pydata.org)：**用于多维数组数据，常用于气候和气象数据集。
- **[Leafmap](https://leafmap.org)：**用最少的代码简化交互式地图的创建。
- **[MapLibre](https://eodagmbh.github.io/py-maplibregl/)：**使用矢量瓦片构建交互式、可定制地图可视化的工具。
- **[WhiteboxTools](https://github.com/jblindsay/whitebox-tools)：**一套用于空间分析的GIS工具。
- **[Geemap](https://geemap.org)：**将Google Earth Engine的强大功能与Python相结合，用于大规模地理空间分析。
- **[Segment-geospatial](https://samgeo.gishub.org)：**用于地理空间分析中图像分割的高级工具。
- **[HyperCoast](https://hypercoast.org)：**用于海岸数据建模和分析。
- **[DuckDB](https://duckdb.org)：**一个快速、可嵌入的分析数据库，具有强大的空间查询能力。
- **[GDAL](https://gdal.org)：**最广泛使用的栅格和矢量数据处理库之一。

这些库将逐步介绍，我们将通过实践练习探索它们的功能。

## 设置Python环境

要跟随本书中的示例和练习，您需要设置一个包含所需地理空间库的Python环境。设置Python环境主要有两种方式：使用**uv**或**conda**。

### uv

[uv](https://github.com/astral-sh/uv)是一个极快的Python包和项目管理器，用Rust编写。它旨在作为pip的直接替代品。

#### 安装uv

您可以按以下方式安装uv：

```bash
# macOS and Linux:
curl -LsSf https://astral.sh/uv/install.sh | sh

# Windows:
powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"

# pip
pip install uv
```

#### 安装geospatial

要安装[geospatial](https://geospatial.gishub.org)包（包含我们将在本书中使用的许多库），运行以下命令：

```bash
uv venv

# macOS and Linux:
source .venv/bin/activate

# Windows:
.venv\Scripts\activate

uv pip install geospatial
uv pip install --find-links https://girder.github.io/large_image_wheels gdal pyproj
```

### pixi

[Pixi](https://pixi.sh)是一个用于以可重现方式安装库和应用程序的包管理工具。它可以从conda-forge通道、PyPI和其他来源安装包。如果您在使用uv安装gdal等地理空间库时遇到问题，可以尝试使用pixi。

#### 安装pixi

您可以按以下方式安装pixi：

```bash
# macOS and Linux:
curl -fsSL https://pixi.sh/install.sh | bash

# Windows:
iwr -useb https://pixi.sh/install.ps1 | iex
```

关闭并重新打开终端或命令提示符，确保pixi命令可用。

#### 安装gdal

要安装gdal和jupyterlab包，请运行以下命令：

```bash
pixi init
pixi add gdal jupyterlab
```

安装包后，您可以打开Jupyter Lab开始处理地理空间数据：

```bash
pixi run jupyter lab
```

### conda

#### 安装Miniconda

Miniconda是Anaconda的轻量级版本，提供管理环境所需的所有核心功能。您可以从官方Miniconda页面下载并安装它：[Miniconda安装指南](https://docs.anaconda.com/miniconda)。

#### 创建新的Conda环境

安装Miniconda后，您可以创建一个专门用于地理空间编程的新环境。这将您的地理空间工具与其他Python项目隔离，有助于避免版本冲突。

在终端或命令提示符中运行以下命令：

```bash
conda create -n geo python=3.11
conda activate geo
```

#### 安装地理空间库

为了更高效地管理多个地理空间库的安装，我们将使用mamba（conda的更快替代方案）。先安装它，然后安装[geospatial](https://geospatial.gishub.org)包，其中包含我们将在本书中使用的许多库：

```bash
conda install -c conda-forge mamba
mamba install -c conda-forge geospatial
```
