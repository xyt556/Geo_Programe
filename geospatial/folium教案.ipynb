### 1.1 Folium 简介（10 分钟）

#### 什么是 Folium？

```
Folium 是一个 Python 库，它将数据可视化能力与 Leaflet.js 地图库结合起来，
让我们可以用 Python 代码生成交互式地图，并以 HTML 格式输出。
```

#### Folium 的核心特点

````python
"""
Folium 核心特点：
1. 基于 Leaflet.js    → 强大的 JavaScript 地图引擎
2. 输出 HTML 文件    → 可在任何浏览器中查看
3. 支持多种瓦片图层  → OpenStreetMap、高德、百度等
4. 丰富的插件系统    → 热力图、聚类、绘图工具等
5. 与 Pandas 集成   → 方便处理地理数据
"""
````

#### 应用场景展示

```
📍 应用场景：
   ✅ 城市交通流量分析
   ✅ 疫情地理分布可视化
   ✅ 商业选址分析
   ✅ 旅游路线规划
   ✅ 气象数据展示
   ✅ 房价热力分布图
```

### 2.2 地图属性配置（20 分钟）

#### 完整参数说明

````python
# ============================================
# 示例 2-4：地图完整参数配置
# ============================================
import folium

m = folium.Map(
    # 核心参数
    location=[39.9042, 116.4074],   # 中心坐标 [lat, lon]
    zoom_start=12,                   # 初始缩放级别
    
    # 地图样式
    tiles='OpenStreetMap',           # 瓦片类型
    
    # 尺寸控制
    width='100%',                    # 地图宽度
    height='100%',                   # 地图高度
    
    # 交互控制
    zoom_control=True,               # 是否显示缩放按钮
    scrollWheelZoom=True,            # 是否允许滚轮缩放
    dragging=True,                   # 是否允许拖动
    
    # 缩放限制
    min_zoom=3,                      # 最小缩放级别
    max_zoom=18,                     # 最大缩放级别
    
    # 其他
    prefer_canvas=True,              # 使用Canvas渲染（性能更好）
    no_touch=False,                  # 是否禁用触摸
    disable_3d=False,                # 是否禁用3D效果
)

m.save('configured_map.html')
m
````

### 2.3 添加比例尺和全屏控件（10 分钟）

````python
# ============================================
# 示例 2-5：添加各种控件
# ============================================
import folium
from folium import plugins

m = folium.Map(location=[39.9042, 116.4074], zoom_start=12)

# 添加比例尺
folium.plugins.MeasureControl(
    position='bottomleft',
    primary_length_unit='kilometers',
    secondary_length_unit='miles',
    primary_area_unit='sqkilometers',
).add_to(m)

# 添加全屏按钮
folium.plugins.Fullscreen(
    position='topright',
    title='全屏',
    title_cancel='退出全屏',
    force_separate_button=True,
).add_to(m)

# 添加鼠标位置显示
folium.plugins.MousePosition(
    position='bottomright',
    separator=' | ',
    prefix='坐标：',
).add_to(m)

m.save('controls_map.html')
m
````

#### 自定义标记图标

### 3.3 批量添加标记（15 分钟）

````python
# ============================================
# 示例 3-6：从 DataFrame 批量添加标记
# ============================================
import folium
import pandas as pd

# 创建示例数据
data = {
    'name': ['故宫', '天坛', '颐和园', '长城', '圆明园', '鸟巢', '水立方', '奥林匹克森林公园'],
    'lat':  [39.9163, 39.8822, 39.9991, 40.4319, 40.0090, 40.0089, 40.0073, 40.0305],
    'lon':  [116.3972, 116.4066, 116.2755, 116.5704, 116.2944, 116.3912, 116.3879, 116.3890],
    'type': ['历史', '历史', '公园', '历史', '历史', '体育', '体育', '公园'],
    'rating': [4.9, 4.8, 4.7, 4.9, 4.6, 4.8, 4.7, 4.6],
    'visitors': [80000, 50000, 40000, 100000, 30000, 60000, 55000, 45000],
}

df = pd.DataFrame(data)

# 创建地图
m = folium.Map(location=[39.9042, 116.4074], zoom_start=11)

# 颜色映射
color_map = {
    '历史': 'red',
    '公园': 'green',
    '体育': 'blue',
}

# 图标映射
icon_map = {
    '历史': 'university',
    '公园': 'tree',
    '体育': 'futbol-o',
}

# 批量添加标记
for _, row in df.iterrows():
    # 创建弹窗 HTML
    popup_html = f"""
    <div style="font-family: Arial; width: 160px;">
        <h4 style="margin: 0; color: #2c3e50;">{row['name']}</h4>
        <hr style="margin: 5px 0;">
        <p style="margin: 2px 0; font-size: 12px;">
            <b>类型：</b>{row['type']}<br>
            <b>评分：</b>{'⭐' * int(row['rating'])} {row['rating']}<br>
            <b>日客流：</b>{row['visitors']:,} 人
        </p>
    </div>
    """
    
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=folium.Popup(
            folium.IFrame(popup_html, width=180, height=130),
            max_width=180
        ),
        tooltip=f"{row['name']} | 评分：{row['rating']}",
        icon=folium.Icon(
            color=color_map[row['type']],
            icon=icon_map[row['type']],
            prefix='fa'
        )
    ).add_to(m)

m.save('batch_markers.html')
print(f"成功添加 {len(df)} 个标记！")
m
````

### 4.1 圆形标记（CircleMarker）（10 分钟）

````python
# ============================================
# 示例 4-1：圆形标记（适合数据大小可视化）
# ============================================
import folium
import pandas as pd
import numpy as np

m = folium.Map(location=[39.9042, 116.4074], zoom_start=11)

# 生成随机数据
np.random.seed(42)
n = 20
data = pd.DataFrame({
    'lat': np.random.uniform(39.8, 40.1, n),
    'lon': np.random.uniform(116.2, 116.6, n),
    'value': np.random.randint(100, 1000, n),
    'category': np.random.choice(['A', 'B', 'C'], n)
})

# 颜色映射
color_map = {'A': '#e74c3c', 'B': '#3498db', 'C': '#2ecc71'}

for _, row in data.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=row['value'] / 50,        # 圆的半径（像素）
        color=color_map[row['category']], # 边框颜色
        fill=True,                        # 是否填充
        fill_color=color_map[row['category']], # 填充颜色
        fill_opacity=0.6,                 # 填充透明度
        weight=2,                         # 边框粗细
        popup=f"类别：{row['category']}<br>数值：{row['value']}",
        tooltip=f"数值：{row['value']}"
    ).add_to(m)

m.save('circle_markers.html')
m
````

### 4.2 折线与多边形（20 分钟）

````python
# ============================================
# 示例 4-2：绘制路线（PolyLine）
# ============================================
import folium

m = folium.Map(location=[39.9042, 116.4074], zoom_start=11)

# 北京旅游路线坐标点
route_points = [
    [39.9163, 116.3972],  # 故宫
    [39.9042, 116.4074],  # 天安门
    [39.8822, 116.4066],  # 天坛
    [39.8650, 116.4700],  # 龙潭公园
    [39.9100, 116.5500],  # 朝阳公园
]

# 标注起点和终点
for i, point in enumerate(route_points):
    if i == 0:
        icon = folium.Icon(color='green', icon='play', prefix='fa')
        label = '出发：故宫'
    elif i == len(route_points) - 1:
        icon = folium.Icon(color='red', icon='stop', prefix='fa')
        label = '终点：朝阳公园'
    else:
        icon = folium.Icon(color='blue', icon='map-marker', prefix='fa')
        label = f'途经点 {i}'
    
    folium.Marker(
        location=point,
        icon=icon,
        tooltip=label
    ).add_to(m)

# 绘制路线
folium.PolyLine(
    locations=route_points,
    color='#e74c3c',       # 线颜色
    weight=4,              # 线宽
    opacity=0.8,           # 透明度
    tooltip='北京一日游路线',
    dash_array=None,       # 虚线样式（如 '5 5'）
).add_to(m)

m.save('route_map.html')
m
````

````python
# ============================================
# 示例 4-3：绘制多边形（Polygon）
# ============================================
import folium

m = folium.Map(location=[39.9042, 116.4074], zoom_start=11)

# 绘制五环（五边形示意）
pentagon_points = [
    [40.0500, 116.3000],
    [40.0800, 116.5000],
    [39.9500, 116.6000],
    [39.8300, 116.4500],
    [39.8500, 116.2500],
]

folium.Polygon(
    locations=pentagon_points,
    color='#3498db',        # 边框颜色
    weight=3,               # 边框粗细
    fill=True,              # 是否填充
    fill_color='#3498db',   # 填充颜色
    fill_opacity=0.2,       # 填充透明度
    tooltip='北京五环区域',
    popup='这是北京五环内大致区域'
).add_to(m)

# 绘制矩形
folium.Rectangle(
    bounds=[[39.88, 116.35], [39.95, 116.45]],  # [[南纬, 西经], [北纬, 东经]]
    color='#e74c3c',
    fill=True,
    fill_color='#e74c3c',
    fill_opacity=0.15,
    tooltip='北京核心区'
).add_to(m)

# 绘制圆形
folium.Circle(
    location=[39.9042, 116.4074],
    radius=5000,            # 半径，单位：米
    color='#2ecc71',
    fill=True,
    fill_color='#2ecc71',
    fill_opacity=0.1,
    tooltip='以天安门为中心5公里范围'
).add_to(m)

m.save('shapes_map.html')
m
````

### 5.1 FeatureGroup 图层分组（20 分钟）

````python
# ============================================
# 示例 5-1：图层分组与控制
# ============================================
import folium

m = folium.Map(location=[39.9042, 116.4074], zoom_start=11)

# 创建不同类别的图层组
layer_history = folium.FeatureGroup(name='🏛️ 历史景点', show=True)
layer_park    = folium.FeatureGroup(name='🌲 公园绿地', show=True)
layer_sport   = folium.FeatureGroup(name='⚽ 体育场馆', show=False)

# 历史景点数据
history_sites = [
    {'name': '故宫',   'lat': 39.9163, 'lon': 116.3972},
    {'name': '天坛',   'lat': 39.8822, 'lon': 116.4066},
    {'name': '颐和园', 'lat': 39.9991, 'lon': 116.2755},
    {'name': '长城',   'lat': 40.4319, 'lon': 116.5704},
]

# 公园数据
parks = [
    {'name': '朝阳公园', 'lat': 39.9393, 'lon': 116.4733},
    {'name': '玉渊潭',   'lat': 39.9050, 'lon': 116.3240},
    {'name': '北海公园', 'lat': 39.9250, 'lon': 116.3920},
]

# 体育场馆数据
sports = [
    {'name': '鸟巢',   'lat': 40.0089, 'lon': 116.3912},
    {'name': '水立方', 'lat': 40.0073, 'lon': 116.3879},
    {'name': '工体',   'lat': 39.9326, 'lon': 116.4528},
]

# 添加历史景点到对应图层
for site in history_sites:
    folium.Marker(
        location=[site['lat'], site['lon']],
        popup=site['name'],
        tooltip=site['name'],
        icon=folium.Icon(color='red', icon='university', prefix='fa')
    ).add_to(layer_history)

# 添加公园到对应图层
for park in parks:
    folium.Marker(
        location=[park['lat'], park['lon']],
        popup=park['name'],
        tooltip=park['name'],
        icon=folium.Icon(color='green', icon='tree', prefix='fa')
    ).add_to(layer_park)

# 添加体育场馆到对应图层
for sport in sports:
    folium.Marker(
        location=[sport['lat'], sport['lon']],
        popup=sport['name'],
        tooltip=sport['name'],
        icon=folium.Icon(color='blue', icon='futbol-o', prefix='fa')
    ).add_to(layer_sport)

# 将图层组添加到地图
layer_history.add_to(m)
layer_park.add_to(m)
layer_sport.add_to(m)

# 添加图层控制器（右上角的图层开关面板）
folium.LayerControl(
    position='topright',
    collapsed=False  # 是否默认折叠
).add_to(m)

m.save('layer_control.html')
m
````

````python
# ============================================
# 示例 6-1：基础热力图
# ============================================
import folium
from folium.plugins import HeatMap
import numpy as np
import pandas as pd

m = folium.Map(location=[39.9042, 116.4074], zoom_start=11)

# 生成模拟人流量数据
np.random.seed(42)
n = 500

# 模拟北京几个热门区域的人流数据
centers = [
    [39.9042, 116.4074, 3.0],  # 天安门（高密度）
    [39.9163, 116.3972, 2.5],  # 故宫
    [39.9200, 116.4358, 2.0],  # 王府井
    [39.8822, 116.4066, 1.5],  # 天坛
]

heat_data = []
for center in centers:
    lat_center, lon_center, intensity = center
    n_points = int(n * intensity / sum(c[2] for c in centers))
    lats = np.random.normal(lat_center, 0.02, n_points)
    lons = np.random.normal(lon_center, 0.02, n_points)
    weights = np.random.uniform(0.5, 1.0, n_points)
    
    for lat, lon, w in zip(lats, lons, weights):
        heat_data.append([lat, lon, w])

# 添加热力图
HeatMap(
    data=heat_data,
    radius=20,              # 每个点的影响半径（像素）
    blur=15,                # 模糊程度
    max_zoom=18,            # 最大缩放级别
    min_opacity=0.3,        # 最小不透明度
    gradient={              # 颜色渐变（0-1之间）
        0.2: 'blue',
        0.4: 'cyan',
        0.6: 'lime',
        0.8: 'yellow',
        1.0: 'red'
    }
).add_to(m)

m.save('heatmap.html')
print(f"热力图已生成，共 {len(heat_data)} 个数据点")
m
````

#### 时序热力图

````python
# ============================================
# 示例 6-2：时序热力图（按时间动态变化）
# ============================================
import folium
from folium.plugins import HeatMapWithTime
import numpy as np

m = folium.Map(location=[39.9042, 116.4074], zoom_start=11)

# 生成 24 小时人流量数据
hours = list(range(24))
time_data = []

for hour in hours:
    # 根据时间模拟不同的人流分布
    if 6 <= hour < 9:    # 早高峰
        intensity = 2.0
        center = [39.9042, 116.4074]
    elif 12 <= hour < 14:  # 午餐时间
        intensity = 1.5
        center = [39.9200, 116.4358]  # 王府井
    elif 17 <= hour < 20:  # 晚高峰
        intensity = 2.5
        center = [39.9042, 116.4074]
    elif 20 <= hour < 23:  # 夜生活
        intensity = 1.8
        center = [39.9350, 116.4550]  # 三里屯
    else:
        intensity = 0.5
        center = [39.9042, 116.4074]
    
    n_points = int(100 * intensity)
    lats = np.random.normal(center[0], 0.02, n_points)
    lons = np.random.normal(center[1], 0.02, n_points)
    weights = np.random.uniform(0.3, 1.0, n_points)
    
    hour_data = [[lat, lon, w] for lat, lon, w in zip(lats, lons, weights)]
    time_data.append(hour_data)

# 添加时序热力图
HeatMapWithTime(
    data=time_data,
    index=[f'{h:02d}:00' for h in hours],  # 时间标签
    radius=25,
    blur=15,
    gradient={0.2: 'blue', 0.6: 'yellow', 1.0: 'red'},
    min_opacity=0.3,
    max_opacity=0.8,
    use_local_extrema=True,
    auto_play=True,          # 自动播放
    display_index=True,      # 显示时间标签
    min_speed=0.5,           # 最小播放速度
    max_speed=5,             # 最大播放速度
    speed_step=0.5,          # 速度步长
    position='bottomleft',   # 控制器位置
    time_slider_drag_update=True,
).add_to(m)

m.save('heatmap_time.html')
m
````

### 7.2 制作 Choropleth 地图（25 分钟）

````python
# ============================================
# 示例 7-2：中国省份 GDP Choropleth 地图
# ============================================
import folium
import pandas as pd
import json
import requests

# 1. 准备统计数据（省份 GDP 数据）
gdp_data = pd.DataFrame({
    'province': ['广东', '江苏', '山东', '浙江', '河南', '四川', 
                 '湖北', '福建', '上海', '北京'],
    'gdp': [12.4, 11.6, 8.3, 7.7, 6.1, 5.4, 
            5.0, 4.9, 4.3, 4.0],  # 万亿元
    'growth_rate': [5.5, 5.8, 5.9, 6.0, 4.1, 6.5, 
                    5.7, 4.7, 3.1, 5.2],  # 增长率%
})

# 2. 加载 GeoJSON（以下用简化示例）
# 实际项目中请从数据源下载真实的省份 GeoJSON
# geojson_url = "https://raw.githubusercontent.com/.../china.json"
# geo_data = requests.get(geojson_url).json()

# 3. 创建地图
m = folium.Map(
    location=[35, 105],   # 中国中心点
    zoom_start=5,
    tiles='CartoDB positron'
)

# 4. 添加 Choropleth 图层
choropleth = folium.Choropleth(
    geo_data='china_provinces.geojson',  # GeoJSON 文件路径或对象
    name='GDP 分布',
    data=gdp_data,
    columns=['province', 'gdp'],          # [key列, value列]
    key_on='feature.properties.name',     # GeoJSON 中的匹配字段
    fill_color='YlOrRd',                  # 颜色方案
    fill_opacity=0.7,
    line_opacity=0.5,
    legend_name='GDP（万亿元）',
    bins=6,                               # 分级数量
    highlight=True,                       # 鼠标悬停高亮
    reset=True
).add_to(m)

# 5. 添加工具提示（显示详细数据）
# 将数据绑定到 GeoJSON 属性
choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(
        fields=['name', 'gdp'],
        aliases=['省份：', 'GDP（万亿）：'],
        style="""
            background-color: white;
            color: #333;
            font-family: Arial;
            font-size: 12px;
            padding: 10px;
            border-radius: 4px;
        """,
        sticky=False
    )
)

folium.LayerControl().add_to(m)
m.save('choropleth_gdp.html')
m
````

### 7.3 GeoJSON 交互效果（10 分钟）

````python
# ============================================
# 示例 7-3：自定义 GeoJSON 样式和交互
# ============================================
import folium
import json

m = folium.Map(location=[35, 105], zoom_start=5)

# 定义样式函数
def style_function(feature):
    """根据属性值返回样式"""
    value = feature['properties'].get('gdp', 0)
    
    # 颜色映射
    if value >= 10:
        color = '#e74c3c'    # 红色：高GDP
    elif value >= 6:
        color = '#e67e22'    # 橙色
    elif value >= 3:
        color = '#f1c40f'    # 黄色
    else:
        color = '#2ecc71'    # 绿色：低GDP
    
    return {
        'fillColor': color,
        'color': 'white',
        'weight': 1.5,
        'fillOpacity': 0.7
    }

# 定义高亮样式函数
def highlight_function(feature):
    """鼠标悬停时的样式"""
    return {
        'fillColor': '#3498db',
        'color': 'white',
        'weight': 3,
        'fillOpacity': 0.9
    }

# 添加 GeoJSON 图层
folium.GeoJson(
    'china_provinces.geojson',
    name='中国省份',
    style_function=style_function,
    highlight_function=highlight_function,
    tooltip=folium.GeoJsonTooltip(
        fields=['name'],
        aliases=['省份：'],
    ),
    popup=folium.GeoJsonPopup(
        fields=['name', 'gdp'],
        aliases=['省份名：', 'GDP：'],
    )
).add_to(m)

folium.LayerControl().add_to(m)
m.save('geojson_interactive.html')
````

### 项目：城市便民设施分布地图

````python
# ============================================
# 综合项目：城市便民设施分布可视化平台
# ============================================

import folium
import pandas as pd
import numpy as np
from folium import plugins

# ============================================
# Step 1：准备数据
# ============================================

np.random.seed(42)

# 城市中心坐标（以成都为例）
CITY_CENTER = [30.5728, 104.0668]

def gen_locations(center, n, std=0.05):
    """生成随机地点坐标"""
    lats = np.random.normal(center[0], std, n)
    lons = np.random.normal(center[1], std, n)
    return list(zip(lats, lons))

# 生成各类设施数据
n = 30  # 每类设施数量

hospitals = pd.DataFrame({
    'name':    [f'第{i+1}人民医院' for i in range(n)],
    'lat':     [loc[0] for loc in gen_locations(CITY_CENTER, n)],
    'lon':     [loc[1] for loc in gen_locations(CITY_CENTER, n)],
    'level':   np.random.choice(['三甲', '三乙', '二甲', '社区'], n),
    'beds':    np.random.randint(200, 2000, n),
    'type':    'hospital'
})

schools = pd.DataFrame({
    'name':    [f'实验{'小' if i%2==0 else '中'}学{i+1}' for i in range(n)],
    'lat':     [loc[0] for loc in gen_locations(CITY_CENTER, n, 0.06)],
    'lon':     [loc[1] for loc in gen_locations(CITY_CENTER, n, 0.06)],
    'level':   np.random.choice(['重点', '普通', '示范'], n),
    'students': np.random.randint(500, 3000, n),
    'type':    'school'
})

parks = pd.DataFrame({
    'name':   [f'城市公园{i+1}' for i in range(n)],
    'lat':    [loc[0] for loc in gen_locations(CITY_CENTER, n, 0.07)],
    'lon':    [loc[1] for loc in gen_locations(CITY_CENTER, n, 0.07)],
    'area':   np.random.randint(1, 100, n),
    'rating': np.round(np.random.uniform(3.5, 5.0, n), 1),
    'type':   'park'
})

metros = pd.DataFrame({
    'name':   [f'{np.random.randint(1,5)}号线{chr(65+i)}站' for i in range(n)],
    'lat':    [loc[0] for loc in gen_locations(CITY_CENTER, n, 0.04)],
    'lon':    [loc[1] for loc in gen_locations(CITY_CENTER, n, 0.04)],
    'line':   np.random.randint(1, 5, n),
    'daily_flow': np.random.randint(5000, 100000, n),
    'type':   'metro'
})

print("数据准备完成！")
print(f"医院：{len(hospitals)} 家")
print(f"学校：{len(schools)} 所")
print(f"公园：{len(parks)} 个")
print(f"地铁站：{len(metros)} 个")

# ============================================
# Step 2：创建地图底图
# ============================================

m = folium.Map(
    location=CITY_CENTER,
    zoom_start=12,
    tiles='CartoDB positron',
)

# 添加多个底图选项
folium.TileLayer('OpenStreetMap', name='街道地图').add_to(m)
folium.TileLayer(
    'https://webrd02.is.autonavi.com/appmaptile?lang=zh_cn&size=1&scale=1&style=8&x={x}&y={y}&z={z}',
    attr='高德地图',
    name='高德地图'
).add_to(m)
folium.TileLayer('CartoDB dark_matter', name='暗色主题').add_to(m)

# ============================================
# Step 3：创建图层组
# ============================================

layer_hospital = folium.FeatureGroup(name='🏥 医疗机构', show=True)
layer_school   = folium.FeatureGroup(name='🎓 学校', show=True)
layer_park     = folium.FeatureGroup(name='🌲 公园绿地', show=True)
layer_metro    = folium.FeatureGroup(name='🚇 地铁站', show=True)
layer_heatmap  = folium.FeatureGroup(name='🔥 人口热力图', show=False)
layer_cluster  = folium.FeatureGroup(name='📍 聚合视图', show=False)

# ============================================
# Step 4：添加医院标记
# ============================================

hospital_colors = {'三甲': 'red', '三乙': 'orange', '二甲': 'blue', '社区': 'green'}

for _, row in hospitals.iterrows():
    popup_html = f"""
    <div style="font-family: Microsoft YaHei; width: 180px;">
        <h4 style="color: #c0392b; margin: 0 0 8px 0;">🏥 {row['name']}</h4>
        <hr style="margin: 0 0 8px 0; border-color: #eee;">
        <table style="font-size: 12px; width: 100%;">
            <tr><td><b>级别：</b></td><td>{row['level']}</td></tr>
            <tr><td><b>床位：</b></td><td>{row['beds']} 张</td></tr>
        </table>
    </div>
    """
    
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=folium.Popup(
            folium.IFrame(popup_html, width=200, height=120),
            max_width=200
        ),
        tooltip=f"🏥 {row['name']} ({row['level']})",
        icon=folium.Icon(
            color=hospital_colors[row['level']],
            icon='hospital-o',
            prefix='fa'
        )
    ).add_to(layer_hospital)

# ============================================
# Step 5：添加学校标记
# ============================================

school_colors = {'重点': 'red', '普通': 'blue', '示范': 'purple'}

for _, row in schools.iterrows():
    popup_html = f"""
    <div style="font-family: Microsoft YaHei; width: 180px;">
        <h4 style="color: #2980b9; margin: 0 0 8px 0;">🎓 {row['name']}</h4>
        <hr style="margin: 0 0 8px 0; border-color: #eee;">
        <table style="font-size: 12px; width: 100%;">
            <tr><td><b>类型：</b></td><td>{row['level']}</td></tr>
            <tr><td><b>学生：</b></td><td>{row['students']:,} 人</td></tr>
        </table>
    </div>
    """
    
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=folium.Popup(
            folium.IFrame(popup_html, width=200, height=120),
            max_width=200
        ),
        tooltip=f"🎓 {row['name']} ({row['level']})",
        icon=folium.Icon(
            color=school_colors[row['level']],
            icon='graduation-cap',
            prefix='fa'
        )
    ).add_to(layer_school)

# ============================================
# Step 6：添加公园标记（圆形）
# ============================================

for _, row in parks.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=max(5, row['area'] / 5),  # 半径与面积成比例
        color='#27ae60',
        fill=True,
        fill_color='#2ecc71',
        fill_opacity=0.6,
        tooltip=f"🌲 {row['name']} | 面积:{row['area']}公顷 | ⭐{row['rating']}",
        popup=f"<b>{row['name']}</b><br>面积：{row['area']} 公顷<br>评分：{'⭐'*int(row['rating'])} {row['rating']}"
    ).add_to(layer_park)

# ============================================
# Step 7：添加地铁站标记
# ============================================

line_colors = {1: '#e74c3c', 2: '#3498db', 3: '#2ecc71', 4: '#9b59b6'}

for _, row in metros.iterrows():
    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=6,
        color=line_colors[row['line']],
        fill=True,
        fill_color=line_colors[row['line']],
        fill_opacity=0.8,
        weight=2,
        tooltip=f"🚇 {row['name']} | 日客流:{row['daily_flow']:,}",
        popup=f"<b>{row['name']}</b><br>{row['line']}号线<br>日均客流：{row['daily_flow']:,} 人次"
    ).add_to(layer_metro)

# ============================================
# Step 8：添加热力图
# ============================================

# 合并所有设施位置生成热力数据
all_locs = (
    list(zip(hospitals['lat'], hospitals['lon'])) +
    list(zip(schools['lat'], schools['lon'])) +
    list(zip(parks['lat'], parks['lon'])) +
    list(zip(metros['lat'], metros['lon']))
)

heat_data = [[loc[0], loc[1], 1.0] for loc in all_locs]

plugins.HeatMap(
    data=heat_data,
    radius=25,
    blur=20,
    gradient={0.2: 'blue', 0.5: 'lime', 0.8: 'orange', 1.0: 'red'},
    min_opacity=0.4,
).add_to(layer_heatmap)

# ============================================
# Step 9：添加聚类视图
# ============================================

cluster = plugins.MarkerCluster().add_to(layer_cluster)

for _, row in pd.concat([hospitals, schools]).iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        tooltip=row['name']
    ).add_to(cluster)

# ============================================
# Step 10：添加所有图层和控件
# ============================================

for layer in [layer_hospital, layer_school, layer_park, 
              layer_metro, layer_heatmap, layer_cluster]:
    layer.add_to(m)

# 图层控制器
folium.LayerControl(collapsed=False, position='topright').add_to(m)

# 全屏按钮
plugins.Fullscreen(position='topleft').add_to(m)

# 鼠标坐标显示
plugins.MousePosition(position='bottomleft').add_to(m)

# 迷你地图
plugins.MiniMap(position='bottomright', tile_layer='CartoDB positron').add_to(m)

# 距离测量工具
plugins.MeasureControl(position='bottomleft').add_to(m)

# ============================================
# Step 11：添加图例
# ============================================

legend_html = """
<div style="
    position: fixed;
    bottom: 60px;
    left: 60px;
    z-index: 1000;
    background: white;
    padding: 12px;
    border-radius: 8px;
    border: 2px solid #ccc;
    box-shadow: 2px 2px 6px rgba(0,0,0,0.3);
    font-family: Microsoft YaHei;
    font-size: 12px;
">
    <h4 style="margin: 0 0 8px 0; color: #2c3e50;">📍 图例</h4>
    <hr style="margin: 0 0 8px 0;">
    <div><span style="color: red;">■</span> 三甲医院</div>
    <div><span style="color: orange;">■</span> 三乙医院</div>
    <div><span style="color: blue;">■</span> 二甲医院/重点学校</div>
    <div><span style="color: green;">■</span> 社区医院/公园</div>
    <div><span style="color: purple;">■</span> 示范学校</div>
    <hr style="margin: 8px 0;">
    <div>🔴 1号线地铁站</div>
    <div>🔵 2号线地铁站</div>
    <div>🟢 3号线地铁站</div>
    <div>🟣 4号线地铁站</div>
</div>
"""

m.get_root().html.add_child(folium.Element(legend_html))

# ============================================
# Step 12：添加标题
# ============================================

title_html = """
<div style="
    position: fixed;
    top: 10px;
    left: 50%;
    transform: translateX(-50%);
    z-index: 1000;
    background: rgba(44, 62, 80, 0.9);
    color: white;
    padding: 10px 20px;
    border-radius: 8px;
    font-family: Microsoft YaHei;
    font-size: 18px;
    font-weight: bold;
    text-align: center;
    box-shadow: 2px 2px 6px rgba(0,0,0,0.4);
">
    🏙️ 城市便民设施分布可视化平台
    <div style="font-size: 12px; font-weight: normal; margin-top: 4px; opacity: 0.8;">
        数据来源：模拟数据 | 更新时间：2024年
    </div>
</div>
"""

m.get_root().html.add_child(folium.Element(title_html))

# ============================================
# Step 13：保存地图
# ============================================

output_file = 'city_facilities_map.html'
m.save(output_file)
print(f"\n✅ 地图已成功保存！文件名：{output_file}")
print(f"📊 数据统计：")
print(f"   - 医院：{len(hospitals)} 家")
print(f"   - 学校：{len(schools)} 所") 
print(f"   - 公园：{len(parks)} 个")
print(f"   - 地铁站：{len(metros)} 个")
print(f"   - 总计：{len(hospitals)+len(schools)+len(parks)+len(metros)} 个设施")
m
````
