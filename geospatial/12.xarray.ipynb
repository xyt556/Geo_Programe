# Xarray

## 概述

[Xarray](https://docs.xarray.dev)是一个强大的Python库，专为处理多维标记数据集而设计，常用于气候科学、海洋学和遥感等领域。它提供了一个高级接口来操作和分析可以看作NumPy数组扩展的数据集。Xarray对于地理空间数据特别有用，因为它支持标记轴（维度）、坐标和元数据，使处理随时间、空间和其他维度变化的数据集变得更加容易。

## 什么是Xarray？

Xarray通过提供标记的多维数组数据结构来扩展NumPy的功能。Xarray中的两个主要数据结构是：

- **DataArray**：一个标记的多维数组，包括维度、坐标和属性。
- **Dataset**：共享相同维度的`DataArray`对象集合。

![](https://docs.xarray.dev/en/stable/_images/dataset-diagram.png)

Xarray对于处理维度具有有意义标签（例如时间、纬度、经度）且元数据很重要的数据集特别有用。

## Xarray数据结构

Xarray提供两个核心数据结构：

1. **DataArray**：一个带有标记维度、坐标和元数据的单个数组。
2. **Dataset**：`DataArray`对象的集合，每个对象对应一个变量，共享相同的维度和坐标。

## 加载数据集

Xarray提供了对多个[教程数据集](https://docs.xarray.dev/en/latest/generated/xarray.tutorial.open_dataset.html)的内置访问，我们可以使用`xr.tutorial.open_dataset`加载。在这里，我们加载一个气温数据集：

```python
ds = xr.tutorial.open_dataset("air_temperature")
ds
```

该数据集以[netCDF](https://www.unidata.ucar.edu/software/netcdf)格式存储，这是科学数据的常见格式。Xarray会自动解析维度和坐标等元数据。

- macOS: `~/Library/Caches/xarray_tutorial_data`

## 使用DataArray

`DataArray`是Xarray中的核心数据结构。它包括数据值、维度（例如时间、纬度、经度）以及每个维度的坐标。

```python
# Access a specific DataArray
temperature = ds["air"]
temperature
```

您也可以使用点表示法访问DataArray：

```python
ds.air
```

## DataArray组件

- **值**：存储在NumPy数组或类似结构中的实际数据。
- **维度**：数据的命名轴（例如时间、纬度、经度）。
- **坐标**：每个维度中值的标签（例如特定时间或地理位置）。
- **属性**：与数据关联的元数据（例如单位、描述）。

您可以提取并打印`DataArray`的值、维度和坐标：

```python
temperature.values
```

```python
temperature.dims
```

```python
temperature.coords
```

```python
temperature.attrs
```

## 索引和选择数据

Xarray允许您根据维度标签轻松选择数据，这在处理地理空间数据时非常直观。

```python
# Select data for a specific time and location
selected_data = temperature.sel(time="2013-01-01", lat=40.0, lon=260.0)
selected_data
```

```python
# Slice data across a range of times
time_slice = temperature.sel(time=slice("2013-01-01", "2013-01-31"))
time_slice
```

## 在DataArray上执行操作

您可以直接在`DataArray`对象上执行算术运算，类似于在NumPy数组上的操作方式。Xarray还会自动处理广播。

```python
# Calculate the mean temperature over time
mean_temperature = temperature.mean(dim="time")
mean_temperature
```

```python
# Subtract the mean temperature from the original data
anomalies = temperature - mean_temperature
anomalies
```

## 使用Xarray进行可视化

Xarray与Matplotlib和其他可视化库集成良好，使得直接从`DataArray`和`Dataset`对象创建绘图变得容易。

```python
# Plot the mean temperature
mean_temperature.plot()
plt.show()
```

您可以通过向`plot`方法传递参数来自定义绘图的外观。例如，您可以指定颜色映射、添加标签和设置图形大小。

```python
mean_temperature.plot(cmap="jet", figsize=(10, 6))
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Mean Temperature")
```

您也可以使用`sel`方法选择特定位置，并绘制该位置的温度时间序列。

```python
# Plot a time series for a specific location
temperature.sel(lat=40.0, lon=260.0).plot()
plt.show()
```

## 高级索引：标签索引与位置索引

Xarray支持标签索引和位置索引，使其在数据选择方面更加灵活。

### 标签索引

您可以使用`.sel()`根据坐标标签选择数据，例如特定时间或位置：

```python
# Select all data from May 2013
ds.sel(time="2013-05")
```

```python
# Slice over time, selecting data between May and July 2013
ds.sel(time=slice("2013-05", "2013-07"))
```

### 位置索引

或者，您可以使用`.isel()`根据坐标的位置选择数据：

```python
# Select the first time step, second latitude, and third longitude
ds.air.isel(time=0, lat=2, lon=3)
```

## 使用Xarray进行高级计算

Xarray提供了几个高级操作，使常见计算变得简单，例如`groupby`、`resample`、`rolling`和`weighted`。

### GroupBy操作

您可以计算数据集的季节平均值等统计数据：

```python
seasonal_mean = ds.groupby("time.season").mean()
seasonal_mean.air.plot(col="season")
```

### 带权重的计算

Xarray允许加权计算，这在网格单元大小不同的地理空间环境中很有用。例如，您可以按单元面积对数据集的平均值进行加权。

```python
cell_area = xr.ones_like(ds.air.lon)  # Placeholder for actual area calculation
weighted_mean = ds.weighted(cell_area).mean(dim=["lon", "lat"])
weighted_mean.air.plot()
```

### 滚动窗口操作

Xarray支持滚动窗口操作，这对于在空间或时间上平滑时间序列数据非常有用。例如，您可以使用5x5窗口在空间上平滑温度数据。

```python
ds.air.isel(time=0).rolling(lat=5, lon=5).mean().plot()
```

同样，您可以使用5天窗口在时间上平滑温度数据。

```python
plt.figure(figsize=(10, 6))
# Select the time series at a specific latitude and longitude
temperature = ds["air"].sel(lat=40.0, lon=260.0)

# Plot the original time series
temperature.plot(label="Original")

# Apply rolling mean smoothing with a window size of 20
smoothed_temperature = temperature.rolling(time=20, center=True).mean()

# Plot the smoothed data
smoothed_temperature.plot(label="Smoothed")

# Add a title and labels
plt.title("Temperature Time Series (lat=40.0, lon=260.0)")
plt.xlabel("Time")
plt.ylabel("Temperature (K)")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

## 读写文件

Xarray支持许多常见的科学数据格式，包括[netCDF](https://www.unidata.ucar.edu/software/netcdf/)和[Zarr](https://zarr.readthedocs.io/)。您可以用几个简单的命令将数据集读写到磁盘。

### 写入netCDF

将数据集保存为netCDF文件：

```python
# Ensure air is in a floating-point format (float32 or float64)
ds["air"] = ds["air"].astype("float32")

# Save the dataset to a NetCDF file
ds.to_netcdf("air_temperature.nc")
```

### 从netCDF读取

从netCDF文件加载数据集：

```python
loaded_data = xr.open_dataset("air_temperature.nc")
loaded_data
```
